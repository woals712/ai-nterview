```python
import csv
import sys
import re

def isHangul(text):
    #Check the Python Version
    pyVer3 =  sys.version_info >= (3, 0)

    if pyVer3 : # for Ver 3 or later
        encText = text
    else: # for Ver 2.x
        if type(text) is not unicode:
            encText = text.decode('utf-8')
        else:
            encText = text
            
    hanCount = len(re.findall(u'[\u3130-\u318F\uAC00-\uD7A3]+', encText))
    return hanCount > 0


f = open('question_data.csv', 'r', encoding='CP949')
fw = open('question_new_data.csv', 'w', encoding='CP949', newline='')
wr = csv.writer(fw)
rdf = csv.reader(f)
for line in rdf:
    for i in range(len(line[0])):
        if isHangul(line[0][i]):
            wr.writerow([line[0][i:], line[1], line[2]])
            print(line[0][i:], 'aa')
            break
f.close()
```

```python
f = open('question_new_data.csv', 'r', encoding='CP949')
rdf = csv.reader(f)
for line in rdf:
    print(line)
f.close()
```
